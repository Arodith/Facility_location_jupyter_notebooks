## Problem dataset

In oder to formulate dataset for facility location we defined a convention way in `NB I - Datasets Generation.ipynb`. By using this convention we will be able to use a set of **helpers** to analyse and 

```python
datasets = {
    "customers": {
        "number": 25,
        "coordinate" : [(i,j) for i in range(5) for j in range(5)],
        "metadata": {
            "color": ["red" for i in range(25)],
            "size":  [10 for i in range(25)],
        }
    },
    "facilities": {
        "number": 16,
        "coordinate" : [(i+0.5,j+0.5) for i in range(4) for j in range(4)],
        "metadata": {
            "color": ["green" for i in range(16)],
            "size":  [20 for i in range(16)],
        }
    }
}
```

```python
import helpers.dataset as hdata
```

```python
d_customersfacilities = hdata.distance_matrices(datasets)['customersfacilities']
```

```python
f_facility = [unif(1,10) for i in range(datasets['facilities']['number'])]
```

```python
I = datasets['customers']['number']
J = datasets['facilities']['number']
```

```python
G = hdata.draw_dataset(datasets)
```

## UFLP

```python
from pkg.uflp_cplex import uflp_cplex

prob, Y, N_icp = uflp_cplex(
            datasets['customers']['number'],
            datasets['facilities']['number'],
            d_customersfacilities,
            f_facility,
            relaxation=False,path="uflp_cplex.lp")
```

```python
G2 = G.copy()

for j,vj in enumerate(Y):
    for i,vi in enumerate(vj):
        if vi == 1:
            G2.add_edge(I+i,j)
```

```python
plt.grid('on')

pos    = nx.get_node_attributes(G2,'pos')
colors = list(nx.get_node_attributes(G2,'color').values())
size   = list(nx.get_node_attributes(G2,'size').values())

nx.draw_networkx(
    G2,pos, node_color=colors,
    node_size=size, with_labels=False
)

plt.savefig("uflp_cplex.png", format="PNG")
```
